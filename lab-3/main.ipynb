### Import the libraries

```python
import pandas as pd
```

### Import the dataset

```python
df = pd.read_csv("iris.csv")
```

```python
print("Dataset Preview:\n", df.head(), "\n")
```

### Box and whisker diagram

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# 1. Box and Whisker Diagram
plt.figure(figsize=(8,6))
sns.boxplot(data=df.drop(columns=["Species","Id"]))
plt.title("Box and Whisker Plot of Iris Features")
plt.show()

```

### Histogram 

```python
# 2. Histogram
df.drop(columns=["Species","Id"]).hist(figsize=(8,6), bins=15)
plt.suptitle("Histogram of Iris Features")
plt.show()
```

### Q-plot


```python
# 3. Q-Q Plot (Quantile-Quantile)
for col in df.drop(columns=["Species","Id"]).columns:
    stats.probplot(df[col], dist="norm", plot=plt)
    plt.title(f"Q-Q Plot of {col}")
    plt.show()
```

### Scatter plot 


```python

# Scatter plots for all numeric columns with hue=Species
sns.pairplot(df[["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm", "Species"]],
             hue="Species", diag_kind="hist", height=2.5)

plt.suptitle("Scatter Plots of Iris Features", y=1.02)
plt.show()
```

### Bar plot

```python
# 5. Bar Plot (species counts)
plt.figure(figsize=(6,4))
sns.countplot(x="Species", data=df)
plt.title("Bar Plot of Species Counts")
plt.show()
```
